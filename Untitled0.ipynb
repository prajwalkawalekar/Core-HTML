```python
''' 5.  Create a description list of full stack web development tech stack, using the <dl> tag. Each term should be
a tech stack name and each description should be a brief explanation of what the tech stack is used for. '''

<!DOCTYPE html>
<html>
<head>
	<title>Full Stack Web Development Tech Stack</title>
</head>
<body>
	<h1>Full Stack Web Development Tech Stack</h1>
	<dl>
		<dt><b>HTML</b></dt>
		<dd>HTML (Hypertext Markup Language) is used for structuring and giving meaning to content on the web.</dd>

		<dt><b>CSS</b></dt>
		<dd>CSS (Cascading Style Sheets) is used for styling and layout of web pages.</dd>

		<dt><b>JavaScript</b></dt>
		<dd>JavaScript is a programming language used for adding interactivity to web pages.</dd>

		<dt><b>React</b></dt>
		<dd>React is a JavaScript library used for building reusable UI components.</dd>

		<dt><b>Node.js</b></dt>
		<dd>Node.js is a JavaScript runtime environment used for running JavaScript on the server-side.</dd>

		<dt><b>Express</b></dt>
		<dd>Express is a Node.js framework used for building web applications.</dd>

		<dt><b>MongoDB</b></dt>
		<dd>MongoDB is a NoSQL database used for storing and retrieving data.</dd>

		<dt><b>RESTful API</b></dt>
		<dd>RESTful API (Application Programming Interface) is an architectural style used for designing networked applications.</dd>

		<dt><b>Git</b></dt>
		<dd>Git is a version control system used for tracking changes in code.</dd>

		<dt><b>GitHub</b></dt>
		<dd>GitHub is a web-based platform used for hosting and sharing code.</dd>
	</dl>
</body>
</html>
```
